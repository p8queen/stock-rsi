```python
#import the packages
import pandas as pd
import numpy as np
from datetime import date
import yfinance as yf
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

today = date.today()
ticker1 = 'SPY' 
start_date = '2014-01-02'
end = today
df = yf.download(ticker1, start=start_date, end=end, interval='1mo')
df
```

```python
# get data using yfinance library
months = []
for i in range(len(df)):
    for j in range(i+1, len(df)):
        if df['Close'][j] >= df['Close'][i] * 1.10:
            tuple_date_i = tuple(df.index[i].date().timetuple()[:3])
            tuple_date_j = tuple(df.index[j].date().timetuple()[:3])
            print('from', tuple_date_i, 'to', tuple_date_j, 'in', j-i, 'months.')
            months.append(j-i)
            break
print('Average months:', np.mean(months))
```

```python
months.sort()
cut_75 = int(len(months) * 0.75)
print('75% out of the data:', months[cut_75:])
```
